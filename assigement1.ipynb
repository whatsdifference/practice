### (a)Answer: 

```python
def get_index(row):
    index=[]
    count=-2
    for x in row:
        count=count+1
        if x==1:
            index.append(count)
    return index
```

```python
#Example
get_index(bob.iloc[0])
```

### (b)Answer: 

```python
def  Jaccard_index(a,b):
    sim=0
    for x in a:
        for y in b:
            if x == y:
                sim=sim+1
    dif=len(a)+len(b)-sim
    jcdx=sim/dif
    return jcdx
```

```python
#Example.1
Jaccard_index([1,2,3],[2,3,4])
```

```python
#Example.2
Jaccard_index([1,2,3,4],[4,5,6,7])
```

### (c)Answer: 

```python
def find_top_two(dataset,arg):
    highest=0
    srdhighest=0
    i1=0
    i2=0
    for index, row in dataset.iterrows():
        ji=Jaccard_index(get_index(row),get_index(arg.iloc[0]))
        if ji>srdhighest:
            srdhighest=ji
            i2=index
            if srdhighest>highest:
                srdh=highest
                highest=srdhighest
                srdhighest=srdh
                i2=i1
                i1=index
    return dataset.iloc[i1],dataset.iloc[i2]
```

```python
find_top_two(purchase_rest,bob)
```

### (d)Answer: 

```python
jac=[]
for index, row in purchase_rest.iterrows():
    jac.append(Jaccard_index(get_index(row),get_index(bob.iloc[0])))
```

```python
print("The row index with the highest Jaccard index is",jac.index(max(jac)),",the set of items are",get_index(purchase_rest.iloc[jac.index(max(jac))]))
print("The highest Jaccard index is",np.max(jac))
```

```python
jac.remove(max(jac))
print("The row index with the second highest Jaccard index is",jac.index(max(jac)),",the set of items are",get_index(purchase_rest.iloc[jac.index(max(jac))]))
print("The second highest Jaccard index is",np.max(jac))
```
